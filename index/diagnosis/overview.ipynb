```python
import pandas as pd
```

```python
DISEASES = [
    "hypertension",
    "diabetes",
    "hyperlipidemia",
    "抑郁",
    "焦虑",
    "不明确的心境障碍",
    "失眠",
    "嗜睡症",
    "睡眠呼吸暂停",
    "pd",
    "发作性睡病",
    "msa",
    "rbd",
]
```

```python
df = pd.DataFrame()
for disease in DISEASES:
    csv = pd.read_csv(f"{disease}.csv")
    label = csv["label"]
    df.loc[disease, "+"] = label.sum()
    df.loc[disease, "-"] = label.size - label.sum()
df["-/+"] = df["-"] / df["+"]
df = df.sort_values(by="-/+")
df
```
